```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,10)
y = 2*x+3
```

```python
plt.plot(x,y)
plt.xlabel('Xaxis')
plt.ylabel('Yaxis')
```

```python
plt.bar(x,y)
plt.xlabel('Xaxis')
plt.ylabel('Yaxis')
```

```python
plt.plot(x,y,label = 'X vs Y')
plt.plot(y,x,label = 'Y vs X')
plt.legend(loc = "upper right")
```
